```python
#Victor
def standardize_string(column: pd.Series):
  clean_column = column.str.lower().str.strip().apply(str)
  return clean_column

def standardize_activity(activity: str):
  """
  Place every activity in a common category based on the words in this activity.
  """
  if activity=='nan':
    return 'unknown'

  activities = ['surfing', 'swimming', 'wading', 'fishing', 'standing', 'spearfishing', 'boarding', 'snorkeling', 'diving', 'walking', 'bathing', 'floating']
  standardized_activity = "other"
  for activ in activities:
    if activ in activity:
      standardized_activity = activ
      break
  #if standardized_activity == "other":
    #print(activity)
  return standardized_activity
df_florida['activity'] = standardize_string(df_florida['activity'])
#df_florida['activity'].value_counts().head(20)
df_florida.shape
```

```python
df_florida['activity'] = df_florida['activity'].apply(standardize_activity)
df_florida['activity'].value_counts().head(10)
```

```python
#Victor

def standardize_type(attack_type: str):
  if attack_type != 'unprovoked' and attack_type != 'provoked':
    return 'unknown'
  return attack_type

df_florida['type'] = standardize_string(df_florida['type'])
df_florida['type'] = df_florida['type'].apply(standardize_type)
#df_florida['type'].value_counts()
df_florida['location'].value_counts().head(20)
```

```python
#Victor
df_florida['location'] = standardize_string(df_florida['location'])

def standardize_location(location: list):
  """
  Return a [location, county] list based on the words in this location.
  """

  counties = ['volusia county', 'brevard county', 'martin county', 'duval county', 'palm beach county', 'monroe county', 'indian river county', 'lee county']
  locations = ['new smyrna beach', 'daytona beach', 'ponce inlet', 'melbourne beach', 'cocoa beach', 'jacksonville beach', 'florida keys', 'juno beach', 'ormond beach', 'palm beach', 'riviera beach']
  loc = "other"
  county = "other"

  if len(location)==1:

    if location[0] == 'nan':

      return ['unknown', 'unknown']

    else:
      loc = 'other'
      county = 'other'
      loca = location[0]

      for l in locations:
        if l in loca:
          loc = l
      for c in counties:
        if c in loca:
          county = c


  elif len(location) ==2:

    loc, county = location[0], location[1]

    for l in locations:
      if l in loc:
        loc = l

    for c in counties:
      if c in loc:
        county = c

  elif len(location) > 2:

    loc, county = location[0], location[-1]

    for l in locations:
      if l in loc:
        loc = l

    for c in counties:
      if c in loc:
        county = c

  return [loc, county]

#df_florida['location'].value_counts().head()

test_df = pd.DataFrame([standardize_location(x.split(',')) for x in df_florida['location']], columns=['precise_location', 'county'])

for i, (loc, county) in enumerate(test_df.values):
  if loc not in ('other', 'unknown') and county in ('other', 'unknown'):

    if loc == 'new smyrna beach' or loc == 'ormond beach' or loc == 'daytona beach': test_df.loc[i, "county"] = 'volusia county'
    if loc == 'florida keys': test_df.loc[i, "county"] = 'monroe county'
    if loc == 'riviera beach' or loc == 'palm beach' or loc == 'juno beach': test_df.loc[i, "county"] = 'palm beach county'

df_florida = pd.concat([df_florida, test_df], axis=1, join='inner')
df_florida.head()
```

```python
#Camil

#Cleaning for sex column
def clear_sex(sex):
    if pd.isna(sex):
        return "F"

    sex = sex.strip().upper()

    if sex == "F":
        return "F"
    elif sex == "N" or sex == "LLI":
        return "F"
    else:
        return "M"


df_florida["sex"] = df_florida["sex"].apply(clear_sex)


#Cleaning for Age column
df_florida["age"] = df_florida['age'].ffill()
df_florida["age"] = df_florida["age"].astype(str).str.extract('(\d+)').astype(float)

df_florida = df_florida.dropna(subset=['age'])

#Add an Age group column
def categorize_age(age):
    if age > 60: return 'Senior'
    if age > 18: return 'Adult'
    if age > 12: return 'Teens'
    else:
      return 'Child'

df_florida['age_group'] = df_florida['age'].apply(categorize_age)


#Date to datetime
def clean_date_prefix(date_value):
  prefixes = ["Reported", "Eearly", "Before", "Fall"]
  for prefix in prefixes:
    if date_value.startswith(prefix):
      return date_value[len(prefix):].strip()
    return date_value

df_florida = df_florida.dropna(subset=['date'])

df_florida['date'] = df_florida['date'].astype(str).apply(clean_date_prefix)

df_florida['date'] = pd.to_datetime(df_florida['date'], format='mixed', dayfirst=True, errors='coerce')
```

```python
# cleaning the data for "Time" into 00h00 format

def standardize_time_format(time_str):
    time_str = str(time_str).strip().lower()

    # Define regex patterns for matching time formats
    patterns = [
        r'(\d{1,2})h(\d{2})',  # matches '12h30', '11h15'
        r'(\d{1,2})h',          # matches '9h', '14h'
        r'(\d{1,2}):(\d{2})',   # matches '11:30', '14:05'
        r'(\d{4})hr',           # matches '1600hr', '1100hr'
        r'(\d{4})',             # matches '1600', '1300' (4 digits)
    ]

    for pattern in patterns:
        match = re.fullmatch(pattern, time_str)
        if match:
            if len(match.groups()) == 2:  # e.g. '12h30'
                hours, minutes = match.groups()
                return f"{int(hours):02}h{int(minutes):02}"
            elif len(match.groups()) == 1:  # e.g. '9h' or '1600hr'
                if 'h' in time_str and 'hr' not in time_str:
                    hours = match.group(1)
                    return f"{int(hours):02}h00"
                else:
                    # Handling 4-digit format like '1600'
                    hours = match.group(1)[:2]
                    minutes = match.group(1)[2:]
                    return f"{int(hours):02}h{int(minutes):02}"

    return 'unknown'
def time_category(time):
  time = str(time)
  if time[0:2].isdigit():
    hour = int(time[0:2])
    if hour > 20 or hour < 7:
      return 'night'
    elif hour >= 12:
      return 'afternoon'
    elif hour >= 7:
      return 'morning'

  return 'unknown'

def clean_and_standardize_time(df, column_name):
    series = df[column_name] #Fixed: Removed extra indent
    series = series[~series.isin(["?", "", np.nan])]

    cleaned_series = series.apply(standardize_time_format)

    #cleaned_series = cleaned_series.dropna()

    df[column_name] = cleaned_series.reset_index(drop=True)
    df[column_name] = df[column_name].apply(time_category)

    #df = df[df[column_name].notna()]

    return df

df_florida = clean_and_standardize_time(df_florida, "time")
df_florida.head()
```

```python
# Function to classify shark species into size groups
def classify_size(species):
    small_species = ['Requiem shark', 'Mako shark']  # Example small species
    medium_species = ['Tiger shark', 'Hammerhead shark']  # Example medium species
    large_species = ['Great white shark', 'Bull shark']  # Example large species
    not_specified = ['Not Specified']

    if species in small_species:
        return 'Small'
    elif species in medium_species:
        return 'Medium'
    elif species in large_species:
        return 'Large'
    elif species in not_specified:
        return 'Unknown'
    else:
        return 'Unknown'

# Apply the size classification function
df_florida['size group'] = df_florida['species'].apply(classify_size)

# Display the updated DataFrame
df_florida.head(20)
```
